```python
import matplotlib.pyplot as plt
%matplotlib inline
data = {
  "sizes": [10, 50, 100, 250, 500, 750, 1000, 1500, 2000, 3000],
  "non-concurrent": [
    0.005,
    0.007,
    0.012,
    0.042,
    0.411,
    4.477,
    11.028,
    40.674,
    104.306,
    179.600
  ],
  "concurrent": [
    0.008,
    0.012,
    0.014,
    0.040,
    0.639,
    2.345,
    5.672,
    20.420,
    53.689,
    201.136
  ]
}

speedup = [ (data["non-concurrent"][i]/data["concurrent"][i]) for i in range(len(data["concurrent"])) ]
print(speedup)
plt.plot(data["sizes"], data["non-concurrent"], 'r')
plt.plot(data["sizes"], data["concurrent"], 'b')
plt.axis([0,4000,0,250])
plt.xlabel('matriz size [n]')
plt.show()

```
